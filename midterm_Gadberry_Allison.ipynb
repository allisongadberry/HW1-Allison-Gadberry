```python
f = open("numbers.txt", "r")

numlist = []


for x in f:
    s = ""
    for c in x:
        if c == ",":
            numlist.append(int(s))
            s = ""
        else:
            s += c
            
print(numlist)
        

```

```python
sortlist = sorted(numlist)
sortlist
```

```python
frequency_dict = {}

for item in sortlist:
    if item in frequency_dict:
        frequency_dict[item] += 1
    else:
        frequency_dict[item] = 1
        
print(frequency_dict)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
x = frequency_dict.keys()
y = frequency_dict.values()
plt.bar(x,y, color="#FF8FAB")
plt.show()
```

```python
with open("midtermdict.json", "w") as jsonfile:
    json.dump(frequency_dict, jsonfile)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Allison_orders.csv")
df.head()
```

```python
df = df.fillna(0)
df.head()
```

```python
df["Total Charged"] = df["Total Charged"].str.replace("$", "").astype(float)
df.head()
```

```python
df.rename(columns = {"Order Date": "Date", "Shipping Address Name": "Name"}, inplace= True)
df["Date"] = pd.to_datetime(df["Date"])
df.head(52)
```

```python
monthly_avgs = df.groupby(df.Date.dt.month)["Total Charged"].mean()
monthly_meds = df.groupby(df.Date.dt.month)["Total Charged"].median()
monthly_mins = df.groupby(df.Date.dt.month)["Total Charged"].min()
monthly_maxs = df.groupby(df.Date.dt.month)["Total Charged"].max()
monthly_stdevs = df.groupby(df.Date.dt.month)["Total Charged"].std()
```

```python
address_dict = {}

for item in df["Name"]:
    if item in address_dict:
        address_dict[item] += 1
    else:
        address_dict[item] = 1
        
print(address_dict)
```

```python
y = address_dict.values()
mycolors = ["#FFC2D1", "#FFFDD2", "#C1E7E3", "#DABFDE"]
plt.pie(y, labels=address_dict.keys(), colors=mycolors)
plt.show
```

```python
plt.plot(monthly_mins, color= "#86BEDA", label="minimum")
plt.plot(monthly_maxs, color= "#838ABD", label="maximum")
plt.plot(monthly_meds, color = "#B3CF99", label="median")
plt.plot(monthly_avgs, color= "#FFC0CB", label="mean")
plt.plot(monthly_stdevs, color= "#F5C77E", label="st. dev")

plt.legend(loc="upper right")
```
