```r
salaries = data.frame(
    id = c(1,2,3,4,5),
    Name = c("Peter", "Amy", "Ryan", "Gary", "Michelle"),
    Salary = c(623.30, 515.20, 611.00, 729.00, 843.25)
)

salaries
```

```r
salaries2 = cbind(salaries, Department = c("IT", "Finance", "HR", "Legal", "Sales"))
salaries2
```

```r
salaries2[c(1,3,5), c(2,3)]
```

```r
x = c(toString(salaries2[1,2]), toString(salaries2[4,2]), toString(salaries2[5,2]))
y = c(salaries2[1,3], salaries2[4,3], salaries2[5,3])

barplot(y, names.arg=x, col="palevioletred1")
```

```r
x = c(min(salaries2[,3]), max(salaries2[,3]), median(salaries2[,3]))
mylabels = c("Minimum", "Maximum", "Median")
mycolors = c("skyblue", "plum", "palegreen")

pie(x, label=mylabels, col=mycolors, main="Salaries Stats")
```
